Zadanie 1

```python
def exp(x):
    return np.exp(-x)
```

```python
def z1(x):
    return (6 - 2*x) / (6 + 4*x + x**2)
```

```python
def z2(x):
    return (6 - 4*x + x**2)/(6 + 2*x)
```

```python
xs = np.linspace(-5,5,1000)
```

```python
plt.figure(figsize=(15, 9))
zs1, zs2, e = z1(xs), z2(xs), exp(xs)
plt.plot(xs, zs1, label="$z_{1}$")
plt.plot(xs, zs2, label="$z_{2}$")
plt.plot(xs, e, label="$exp(-x)$")
plt.legend(loc="best")
plt.title("Przybliżenia Padego $z_1$ i $z_2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres1.png")
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.subplot(1,2,1)
plt.plot(xs, zs1, label="$z_{1}$")
plt.plot(xs, e, label="$exp(-x)$")
plt.title("Przybliżenie Padego $z_1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.plot(xs, zs2, label="$z_{2}$")
plt.plot(xs, e, label="$e(-x)$")
plt.title("Przybliżenie Padego $z_2$")
plt.legend(loc="best")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres2.png")
plt.show()
```

```python
error_z1 = np.abs(e - zs1)
error_z2 = np.abs(e - zs2)
```

```python
plt.figure(figsize=(15, 9))
plt.plot(xs, error_z1, label="Błąd z1", linestyle=":")
plt.plot(xs, error_z2, label="Błąd z2", linestyle="--")
plt.title("Błędy przybliżeń Padé")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres3.png")
plt.show()
```

Błąd zaokrąglenia występuje w obliczeniach komputerowych z powodu ograniczeń w sposobie, w jaki komputery przechowują liczby zmiennoprzecinkowe. Komputery wykorzystują system binarny do reprezentowania liczb, co oznacza, że liczby są zapisywane jako sekwencje zer i jedynek. Jednak nie wszystkie liczby dziesiętne mogą być dokładnie zapisane w systemie binarnym, podobnie jak niektóre liczby ułamkowe nie mogą być dokładnie zapisane w systemie dziesiętnym 

Liczby zmiennoprzecinkowe są zapisywane w standardzie IEEE 754, który przydziela stałą ilość bitów do przechowywania mantysy (części ułamkowej) i wykładnika (części wskazującej skalę). Oznacza to, że tylko ograniczona liczba cyfr może zostać dokładnie zapisana. Gdy liczba ma więcej cyfr niż dostępna pamięć, nadmiar cyfr jest "obcinany", co powoduje zaokrąglenie liczby.

W przypadku wyrażenia 7/100 * 100 - 7 liczba 7/100 (czyli 0.07) nie jest dokładnie reprezentowana w systemie binarnym, co prowadzi do minimalnego błędu, który zostaje powielony przy dalszych operacjach. Choć dla nas ten błąd jest praktycznie niewidoczny (bo wynik jest bliski zeru), komputer zapisuje go jako bardzo małą wartość, np. 
8.88 * 10^(-16)

Liczby takie jak 7, 14, 28, 29 generują takie same błędy zaokrąglenia przy operacjach z ułamkami dziesiętnymi. Dzieje się tak, ponieważ ich reprezentacja w systemie binarnym wymaga przybliżenia (zaokrąglenia) na pewnym etapie.

Zadanie 4

Mamy 
$$
X = 0 | 01111111|10110011001100110011001
$$
Czyli
$$
X_{-} = (-1)^0 \cdot 2^{(1+2+4+8+16+32+64 - 127)} \cdot 1,10110011001100110011001_2 = 1,10110011001100110011001_2
$$
Wówczas
$$
X_+ = X_{-} + 2^{-23} = 1,10110011001100110011010_2
$$
Obliczmy różnice
$$
X - X_{-} = 0,10011001..._2 \cdot 2^{-24}
$$

```python
rozwiniecie_wieksza = "10110011001100110011010"
rozwiniecie_mniejsza = "10110011001100110011001"
wartosc = 1.7
wieksza = 1 + sum(int(rozwiniecie_wieksza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_wieksza)))
mniejsza = 1 + sum(int(rozwiniecie_mniejsza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_mniejsza)))
roznica_wieksza = np.absolute(wartosc - wieksza)
roznica_mniejsza = np.absolute(wartosc - mniejsza)
print(roznica_wieksza, roznica_mniejsza)
```

Widzimy, że róznica między wartością dokładną $X$ a $X_{+}$ jest mniejsza niż różnica między $X$ i $X_{-}$, a zatem przyjmujemy $fl(X) = X_{+}$.

Wówczas błąd bezwzględny jest równy
$$
|fl(X) - X| = |X_{+} - X| = 4.768371586472142 \cdot 10^{-8}
$$
Obliczmy błąd względny

```python
blad_wzgledny = roznica_wieksza / wartosc
print(blad_wzgledny)
```

Czyli błąd względny wynosi
$$
\frac{|fl(X) - X|}{|X|} = 2.804924462630672 \cdot 10^{-8}
$$

Zadanie 5

Pierwsza inmplementacja (uogólniona)

```python
def wielomian1(poly, n, x):
    potega = n - 1
    wynik = 0
    for wspolczynnik in poly:
        wynik += wspolczynnik * x ** potega
        potega -= 1
    return wynik
```

```python
poly = [6, 5, -13, 1, 1]
n = len(poly)
```

```python
xs = np.linspace(-10, 10, 200000)
```

```python
def main():
    for x in xs:
        wielomian1(poly, n, x)

if __name__ == '__main__':
    cProfile.run('main()')
```

Druga implementacja (szczególna)

```python
def wielomian2(x):
    return 6 * x**4 + 5 * x**3 - 13 * x**2 + x + 1
```

```python
def main():
    for x in xs:
        wielomian2(x)

if __name__ == '__main__':
    cProfile.run('main()')
```

```python
def horner1(wspolczynniki, n , x):
    wynik = wspolczynniki[0]
    ind = 0
    while ind < n:
        wynik = wynik * x + wspolczynniki[ind]
        ind += 1
    return wynik
```
