Zadanie 1

```python
def exp(x):
    return np.exp(-x)
```

```python
def z1(x):
    return (6 - 2*x) / (6 + 4*x + x**2)
```

```python
def z2(x):
    return (6 - 4*x + x**2)/(6 + 2*x)
```

```python
xs = np.linspace(-5,5,1000)
```

```python
plt.figure(figsize=(15, 9))
zs1, zs2, e = z1(xs), z2(xs), exp(xs)
plt.plot(xs, e, color = "b", label="$exp(-x)$", linewidth="0.7")
plt.plot(xs, zs1, color="g", label="$z_{1}$", linestyle="--")
plt.plot(xs, zs2, color ="r", label="$z_{2}$", linestyle=":")
plt.legend(loc="best")
plt.title("Przyblienia Padego $z_1$ i $z_2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres1.png")
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.subplot(1,2,1)
plt.plot(xs, zs1, color="g", linestyle="--", label="$z_{1}$")
plt.plot(xs, e, label="$exp(-x)$", linewidth="1")
plt.title("Przyblienie Padego $z_1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.plot(xs, zs2, color ="r", linestyle =":",label="$z_{2}$")
plt.plot(xs, e, label="$exp(-x)$", linewidth="1")
plt.title("Przyblienie Padego $z_2$")
plt.legend(loc="best")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres2.png")
plt.show()
```

```python
error_z1 = np.abs(e - zs1)
error_z2 = np.abs(e - zs2)
```

```python
plt.figure(figsize=(15, 9))
plt.plot(xs, error_z1, label="Bd $z_1$", color="g", linestyle="--")
plt.plot(xs, error_z2, label="Bd $z_2$", color="r", linewidth=2, linestyle=":")
plt.title("Bdy przyblie Padego")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres3.png")
plt.show()
```

Bd zaokrglenia wystpuje w obliczeniach komputerowych z powodu ogranicze w sposobie, w jaki komputery przechowuj liczby zmiennoprzecinkowe. Komputery wykorzystuj system binarny do reprezentowania liczb, co oznacza, e liczby s zapisywane jako sekwencje zer i jedynek. Jednak nie wszystkie liczby dziesitne mog by dokadnie zapisane w systemie binarnym, podobnie jak niektre liczby uamkowe nie mog by dokadnie zapisane w systemie dziesitnym 

Liczby zmiennoprzecinkowe s zapisywane w standardzie IEEE 754, ktry przydziela sta ilo bitw do przechowywania mantysy (czci uamkowej) i wykadnika (czci wskazujcej skal). Oznacza to, e tylko ograniczona liczba cyfr moe zosta dokadnie zapisana. Gdy liczba ma wicej cyfr ni dostpna pami, nadmiar cyfr jest "obcinany", co powoduje zaokrglenie liczby.

W przypadku wyraenia 7/100 * 100 - 7 liczba 7/100 (czyli 0.07) nie jest dokadnie reprezentowana w systemie binarnym, co prowadzi do minimalnego bdu, ktry zostaje powielony przy dalszych operacjach. Cho dla nas ten bd jest praktycznie niewidoczny (bo wynik jest bliski zeru), komputer zapisuje go jako bardzo ma warto, np. 
8.88 * 10^(-16)

Liczby takie jak 7, 14, 28, 29 generuj takie same bdy zaokrglenia przy operacjach z uamkami dziesitnymi. Dzieje si tak, poniewa ich reprezentacja w systemie binarnym wymaga przyblienia (zaokrglenia) na pewnym etapie.

Zadanie 4

Mamy 
$$
X = 0 | 01111111|10110011001100110011001
$$
Czyli
$$
X_{-} = (-1)^0 \cdot 2^{(1+2+4+8+16+32+64 - 127)} \cdot 1,10110011001100110011001_2 = 1,10110011001100110011001_2
$$
Wwczas
$$
X_+ = X_{-} + 2^{-23} = 1,10110011001100110011010_2
$$
Obliczmy rnice
$$
X - X_{-} = 0,10011001..._2 \cdot 2^{-24}
$$

```python
rozwiniecie_wieksza = "10110011001100110011010"
rozwiniecie_mniejsza = "10110011001100110011001"
wartosc = 1.7
wieksza = 1 + sum(int(rozwiniecie_wieksza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_wieksza)))
mniejsza = 1 + sum(int(rozwiniecie_mniejsza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_mniejsza)))
roznica_wieksza = np.absolute(wartosc - wieksza)
roznica_mniejsza = np.absolute(wartosc - mniejsza)
print(roznica_wieksza, roznica_mniejsza)
```

Widzimy, e rznica midzy wartoci dokadn $X$ a $X_{+}$ jest mniejsza ni rnica midzy $X$ i $X_{-}$, a zatem przyjmujemy $fl(X) = X_{+}$.

Wwczas bd bezwzgldny jest rwny
$$
|fl(X) - X| = |X_{+} - X| = 4.768371586472142 \cdot 10^{-8}
$$
Obliczmy bd wzgldny

```python
blad_wzgledny = roznica_wieksza / wartosc
print(blad_wzgledny)
```

Czyli bd wzgldny wynosi
$$
\frac{|fl(X) - X|}{|X|} = 2.804924462630672 \cdot 10^{-8}
$$

Zadanie 5

Pierwsza implementacja (uoglniona)

```python
def wielomian1(wspolczynniki, n, x):
    potega = n - 1
    wynik = 0
    for wspolczynnik in wspolczynniki:
        wynik += wspolczynnik * x ** potega
        potega -= 1
    return wynik
```

```python
wspolczynniki = [6, 5, -13, 1, 1]
n = len(wspolczynniki)
```

```python
xs = np.linspace(-10, 10, 200000)
```

```python
def main():
    for x in xs:
        wielomian1(wspolczynniki, n, x)

if __name__ == '__main__':
    cProfile.run('main()')
```

Druga implementacja (szczeglna)

```python
def wielomian2(x):
    return 6 * x**4 + 5 * x**3 - 13 * x**2 + x + 1
```

```python
def main():
    for x in xs:
        wielomian2(x)

if __name__ == '__main__':
    cProfile.run('main()')
```

Zadanie 6

Implementacja pierwsza (uoglniona)

```python
def horner1(wspolczynniki, n , x):
    wynik = wspolczynniki[0]
    ind = 0
    while ind < n:
        wynik = wynik * x + wspolczynniki[ind]
        ind += 1
    return wynik
```

```python
wspolczynniki = [1, 1, -13, 5, 6]
n = len(wspolczynniki)
```

```python
def main():
    for x in xs:
        horner1(wspolczynniki, n, x)
```

```python
if __name__ == '__main__':
    cProfile.run('main()')
```
