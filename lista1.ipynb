Zadanie 1

```python
def exp(x):
    return np.exp(-x)
```

```python
def z1(x):
    return (6 - 2*x) / (6 + 4*x + x**2)
```

```python
def z2(x):
    return (6 - 4*x + x**2)/(6 + 2*x)
```

```python
xs = np.linspace(-5,5,1000)
```

```python
plt.figure(figsize=(15, 9))
zs1, zs2, e = z1(xs), z2(xs), exp(xs)
plt.plot(xs, zs1, label="$z_{1}$")
plt.plot(xs, zs2, label="$z_{2}$")
plt.plot(xs, e, label="$exp(-x)$")
plt.legend(loc="best")
plt.title("Przybli≈ºenia Padego $z_1$ i $z_2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres1.png")
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.subplot(1,2,1)
plt.plot(xs, zs1, label="$z_{1}$")
plt.plot(xs, e, label="$exp(-x)$")
plt.title("Przybli≈ºenie Padego $z_1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.plot(xs, zs2, label="$z_{2}$")
plt.plot(xs, e, label="$e(-x)$")
plt.title("Przybli≈ºenie Padego $z_2$")
plt.legend(loc="best")
plt.tight_layout()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres2.png")
plt.show()
```

```python
error_z1 = np.abs(e - zs1)
error_z2 = np.abs(e - zs2)
```

```python
plt.figure(figsize=(15, 9))
plt.plot(xs, error_z1, label="B≈ÇƒÖd z1", linestyle=":")
plt.plot(xs, error_z2, label="B≈ÇƒÖd z2", linestyle="--")
plt.title("B≈Çƒôdy przybli≈ºe≈Ñ Pad√©")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig(r"C:\Users\Rafal\OneDrive\Dokumenty\GitHub\Metody-numeryczne\wykres3.png")
plt.show()
```

B≈ÇƒÖd zaokrƒÖglenia wystƒôpuje w obliczeniach komputerowych z powodu ogranicze≈Ñ w sposobie, w jaki komputery przechowujƒÖ liczby zmiennoprzecinkowe. Komputery wykorzystujƒÖ system binarny do reprezentowania liczb, co oznacza, ≈ºe liczby sƒÖ zapisywane jako sekwencje zer i jedynek. Jednak nie wszystkie liczby dziesiƒôtne mogƒÖ byƒá dok≈Çadnie zapisane w systemie binarnym, podobnie jak niekt√≥re liczby u≈Çamkowe nie mogƒÖ byƒá dok≈Çadnie zapisane w systemie dziesiƒôtnym 

Liczby zmiennoprzecinkowe sƒÖ zapisywane w standardzie IEEE 754, kt√≥ry przydziela sta≈ÇƒÖ ilo≈õƒá bit√≥w do przechowywania mantysy (czƒô≈õci u≈Çamkowej) i wyk≈Çadnika (czƒô≈õci wskazujƒÖcej skalƒô). Oznacza to, ≈ºe tylko ograniczona liczba cyfr mo≈ºe zostaƒá dok≈Çadnie zapisana. Gdy liczba ma wiƒôcej cyfr ni≈º dostƒôpna pamiƒôƒá, nadmiar cyfr jest "obcinany", co powoduje zaokrƒÖglenie liczby.

W przypadku wyra≈ºenia 7/100 * 100 - 7 liczba 7/100 (czyli 0.07) nie jest dok≈Çadnie reprezentowana w systemie binarnym, co prowadzi do minimalnego b≈Çƒôdu, kt√≥ry zostaje powielony przy dalszych operacjach. Choƒá dla nas ten b≈ÇƒÖd jest praktycznie niewidoczny (bo wynik jest bliski zeru), komputer zapisuje go jako bardzo ma≈ÇƒÖ warto≈õƒá, np. 
8.88 * 10^(-16)

Liczby takie jak 7, 14, 28, 29 generujƒÖ takie same b≈Çƒôdy zaokrƒÖglenia przy operacjach z u≈Çamkami dziesiƒôtnymi. Dzieje siƒô tak, poniewa≈º ich reprezentacja w systemie binarnym wymaga przybli≈ºenia (zaokrƒÖglenia) na pewnym etapie.

$$
W√≥wczas
$$
X_+ = X_{-} + 2^{-23} = 1,10110011001100110011010_2
$$
Obliczmy r√≥≈ºnice
$$
X - X_{-} = 0,10011001..._2 \cdot 2^{-24}
$$

```python
rozwiniecie_wieksza = "10110011001100110011010"
rozwiniecie_mniejsza = "10110011001100110011001"
wartosc = 1.7
wieksza = 1 + sum(int(rozwiniecie_wieksza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_wieksza)))
mniejsza = 1 + sum(int(rozwiniecie_mniejsza[i]) * 2**-(i + 1) for i in range(len(rozwiniecie_mniejsza)))
roznica_wieksza = np.absolute(wartosc - wieksza)
roznica_mniejsza = np.absolute(wartosc - mniejsza)
print(roznica_wieksza, roznica_mniejsza)
```

$$

$$

```python
blad_wzgledny = roznica_wieksza / wartosc
print(blad_wzgledny)
```

Zadanie 5

Pierwsza inmplementacja (uog√≥lniona)

```python
def wielomian1(poly, n, x):
    potega = n - 1
    wynik = 0
    for wspolczynnik in poly:
        wynik += wspolczynnik * x ** potega
        potega -= 1
    return wynik
```

```python
poly = [6, 5, -13, 1, 1]
n = len(poly)
```

```python
xs = np.linspace(-10, 10, 200000)
```

```python
def main():
    for x in xs:
        wielomian1(poly, n, x)

if __name__ == '__main__':
    cProfile.run('main()')
```

Druga implementacja (szczeg√≥lna)

```python
def wielomian2(x):
    return 6 * x**4 + 5 * x**3 - 13 * x**2 + x + 1
```

```python
def main():
    for x in xs:
        wielomian2(x)

if __name__ == '__main__':
    cProfile.run('main()')
```

```python
def horner1(wspolczynniki, n , x):
    wynik = wspolczynniki[0]
    ind = 0
    while ind < n:
        wynik = wynik * x + wspolczynniki[ind]
        ind += 1
    return wynik
```
